# Задание 2

```python
#Перевожу формат 2 необходимых столбцов из object в datetime

orders.order_delivered_customer_date = pd.to_datetime(orders.order_delivered_customer_date)
orders.order_estimated_delivery_date = pd.to_datetime(orders.order_estimated_delivery_date)
```

```python
#Создаю столбцы,в которых будут храниться год и месяц предполагаемой доставки

orders['month_of_delivery'] = orders.order_estimated_delivery_date.dt.month
orders['year_of_delivery'] = orders.order_estimated_delivery_date.dt.year
```

```python
#Создаю столбец "не доставлено", в котором значение 1 означает, что статус "доставлено" у товара не появилось, а 0, что он все же был доставлен

orders['not_delivered'] = orders.apply(lambda x: 0 if x.order_status == 'delivered' else 1, axis=1)
```

```python
#Сперва группирую данные по месяцу и году предполагаемой доставки и по статусу заказа, после чего суммирую колонку "не доставлено". Затем еще раз группирую по столбцу статуса заказа и считаю среднее по всем месяцам

orders.groupby(['year_of_delivery', 'month_of_delivery', 'order_status'], as_index = False).not_delivered.sum()\
    .groupby('order_status', as_index = False).not_delivered.mean().query('order_status != "delivered"')
```

По итогу анализа получены следующие результаты. В среднем в месяц по причине *отмены заказа* не доставляются приблизительно **24.04 заказа**, из-за *недоступности* - **29 заказов**, остальные же заказы не были доставлены на различных этапах формирования и доставки заказа

# Задание 4

```python
#Перевожу необходиые столбцы в формат datetime в объединенной таблицы. Ранее по заданиям такой необходимости особо не было

orders_with_id.order_delivered_customer_date = pd.to_datetime(orders_with_id.order_delivered_customer_date)
orders_with_id.order_approved_at = pd.to_datetime(orders_with_id.order_approved_at)
orders_with_id.order_estimated_delivery_date = pd.to_datetime(orders_with_id.order_estimated_delivery_date)
```

```python
#Создаю столбцы с месяцем и годом подтверждения заказа, а также с расчетом числа недель в одном месяце 

orders_with_id['order_month'] = orders_with_id.order_approved_at.dt.month
orders_with_id['order_year'] = orders_with_id.order_approved_at.dt.year
orders_with_id['weeks_in_month'] = orders_with_id.order_approved_at.dt.daysinmonth / 7
```

```python
#Группирую совершенные покупки по id пользователя, году, месяцу и числу недель в месяце и считаю количество покупок для каждой комбинации

orders_per_month = orders_with_id.query('order_status == "delivered"').groupby(['customer_unique_id','order_year', 'order_month', 'weeks_in_month'], as_index = False).order_id.count()
```

```python
#Создаю столбец "заказов в неделю"

orders_per_month['orders_per_week'] = orders_per_month.order_id / orders_per_month.weeks_in_month
```

```python
#Удаляю столбцы "недель в месяце" и "id заказа" за ненадобностью

orders_per_month = orders_per_month.drop(['weeks_in_month', 'order_id'], axis = 1)
```

```python
#Определяю количество покупок в неделю, для удобства сортируя данные по убывонию

orders_per_month.sort_values('orders_per_week', ascending = False)
```

# Задание 5

В рамках данного задания я определяю когорты людей по дате (месяц и год) их первой покупки, так как данных о дате регистрации в таблицах не имеется

Также я буду рассматривать данные за все года, т.е. за каждый декабрь и январь, что представлены в таблицах. Исхожу из того, что самые поздние даты в таблицах затрагивают сентябрь-октябрь 2018 года, так что предположение, что просят изучить именно последние 2 зимних месяца кажутся мне немного странными. Таким образом, в рамках задания я буду исходить из предположения, что я изучаю динамику изменений по годам в, теоретически, самый прибыльный (декабрь) и в самый неприбыльный (январь) месяцы

```python
#Для удобства создаю таблицу, где оставляю только доставленные заказы

del_orders_with_id = orders_with_id.query('order_status == "delivered"')
```

```python
#Сперва создаю таблицу, в которой будут находиться только самые первые заказы пользователей

first_orders_index = del_orders_with_id.groupby('customer_unique_id').order_approved_at.idxmin().dropna()
first_order = del_orders_with_id.loc[first_orders_index]
```

```python
first_order
```

```python
#Создаю колонку "когорта", в которой делю пользователей по месяцу и году первой совершенной покупки

first_order['cohort'] = first_order.apply(lambda x: '.'.join([str(int(x.order_month)), str(int(x.order_year))]), axis = 1)
```

```python
#Объединяю таблицу с отфильтрованными заказами и таблицы с когортами, добавляя их в изначальную таблицу

cohorted_users = del_orders_with_id.merge(first_order[['customer_unique_id','cohort']], how = 'left', on = 'customer_unique_id')
```

```python
#Отфильтровываю данные так, чтобы остались только данные за заказы, подтвержденные в декабре или январе

filtered_orders = cohorted_users.loc[(cohorted_users.order_approved_at.dt.month == 1) | (cohorted_users.order_approved_at.dt.month == 12)]
```

```python
#Создаю столбец с retention на 3 месяц, где 1 - пользователь совершил повторную покупку cпустя 90 дней, 0 - не совершил

filtered_orders['ret_month_3'] = filtered_orders.groupby('customer_unique_id').order_approved_at.transform(
                                        lambda x: (x - x.min()).dt.days >= 90).astype(int)
```

```python
#Создаю сводную таблицу, дабы посмотреть на результаты

filtered_orders.pivot_table(values = 'ret_month_3', index = 'cohort', columns = filtered_orders.order_approved_at.dt.month, aggfunc='sum', fill_value = 0)
```

Самым высоким показателем retention по истечению 3 месяцев с первой покупки оказались пользователи, совершившие первую покупку в январе 2017 года

# Задание 6

```python
#Создаю таблицу, в которой хранятся данные о RFM пользователей. Для начала сохраняю уникальные id пользователей


RFM_table = pd.DataFrame()
```

```python
#В переменную today сохраняю последнюю дату, которая обозначена в таблице

today = del_orders_with_id.order_approved_at.max()
```

```python
#Вычисляю количество дней от подтверждения заказа до последней даты в таблице

del_orders_with_id['days_since_last'] = del_orders_with_id.order_approved_at.dropna().drop_duplicates()\
                                        .apply(lambda x: (today - x).days)
```

```python
#Создаю таблицу, в которую переношу id пользователей и число дней с заказа

df = del_orders_with_id[['customer_unique_id','days_since_last']]
```

```python
#Создаю таблицу RFM, оставляя только уникальны id пользователей и наименьшее из значений дней с последнего заказа

RFM_table = df.groupby('customer_unique_id', as_index = False).days_since_last.agg(lambda x: x.min())
```

```python
#В изначальной таблице создаю столбец "frequency", считая количество заказов, совершенных одним id. Т.к. дупликатов по столбцу order_id нет, результат должен оказаться верен

del_orders_with_id['frequency'] = del_orders_with_id.groupby('customer_unique_id').order_id.transform('count')
```

```python
#Добавляю к исходной таблице столбец с ценами из таблицы о товарах

del_orders_with_id = del_orders_with_id.merge(items[['order_id', 'price']], how = 'left', on = 'order_id')
```

```python
#Создаю в таблице столбец "monetary", просуммировав уплаченную сумму по всем заказам

monetary_summary = del_orders_with_id.groupby('customer_unique_id', as_index=False)['price'].sum()
monetary_summary.rename(columns={'price': 'monetary'}, inplace=True)
del_orders_with_id = del_orders_with_id.merge(monetary_summary, on='customer_unique_id', how='left')
```

```python
#Добавляем monetary и frequency в таблицу, переименовываем столбец, удаляем NaN и дупликаты

RFM_table = RFM_table.rename(columns = {'days_since_last': 'recency'})\
                     .merge(del_orders_with_id[['customer_unique_id', 'frequency', 'monetary']], how = 'left')\
                     .drop_duplicates().dropna()
```

```python
#Изучаем статистику по определенным критериям

RFM_table.describe()
```

Теперь необходимо определить, на какие кластеры я буду делить пользователей. Всего в таблице находится 85627 человек, так что дробление на 3 кластера считаю недостаточным. Деление же на 4 кластера кажутся мне более, чем достаточными

**Recency** - имеются некоторые выбросы, так что самым разумным вариантом для меня кажется взятие кластеров по процентилям (меньше 25, 25-50, 50-75 и больше 75)

**Frenquency** - разброс данных слишком большой, распределение данных слишком неравномерное, так что кластеры придется подбирать не по процентилям.
Самое частое значение - **1**, так что определяю его в отдельный кластер. Покупки размером в **2** объекта, как мне кажется, также можно считать отдельным кластером, поскольку человека, совершившего 2 покупки, еще нельзя посчитать постоянным покупателем, и к тому же доля людей, совершивших именно 2 закупки, больше, чем совершивших от 3 в несколько раз. Людей, купивших **от 3 до 6** объектов, я определяю в третий кластер, поскольку это уже достаточно большое (относительно) число закупок, но даже 6 закупок в сумме совершило больше людей, чем всего суммарно закупилось от 7. Таким образом, последний кластер занимают люди, закупившиеся **больше 6 раз**

**Monetary** - распределение не является нормальным, имеются сильные выбросы, так что разумным считаю использование класеров по процентилям, аналогично Recency

```python
RFM_table['R'] = pd.qcut(RFM_table.recency, q=4, labels=[4, 3, 2, 1])
RFM_table['F'] = pd.cut(RFM_table.frequency, bins=[0, 1, 2, 6, float('inf')], labels=[1, 2, 3, 4])
RFM_table['M'] = pd.qcut(RFM_table.monetary, q=4, labels=[1, 2, 3, 4])
RFM_table['RFM'] = RFM_table.R.astype(str) + RFM_table.F.astype(str) + RFM_table.M.astype(str)
```
